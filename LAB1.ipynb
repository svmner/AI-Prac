```python
#Q1

class Queue_Two_Stacks:
    def __init__(self):
        self.stack1=[]
        self.stack2=[]
    
    def enqueue(self,item):
        self.stack1.append(item)
        
    def dequeue(self):
        if len(self.stack2) ==0:
            if len(self.stack1) == 0:
                print("Empty Queue!!")
            
            while len(self.stack1) > 0:
                last_stack1_item = self.stack1.pop()
                self.stack2.append(last_stack1_item)
                
        return self.stack2.pop()

if __name__ == "__main__" :
    
    q = Queue_Two_Stacks()
    
    
    n = int(input("Enter no of elements"))
    
    print("Enter integer elements")
    
    for i in range(n):
        q.enqueue(int(input()))
        
    print("Your queue is: ")
    print(q.stack1)
    
    print("Dequeueing...")
    for i in range(n):
        print(q.dequeue())
        
```

```python
#Q4

class Node:
    def __init__(self, v):
        self.left = None
        self.right = None
        self.data = v
        
def printInorder(root):
    if root:
        printInorder(root.left)
        print(root.data,end=" ")
        printInorder(root.right)
            
def printPreOrder(node):
    if node is None:
        return
    print(node.data, end = " ")
    printPreOrder(node.left)
    printPreOrder(node.right)
    
def printPostOrder(node):
    if node is None:
        return
    
    printPostOrder(node.left)
    printPostOrder(node.right)
    print(node.data, end = " ")   
 

if __name__ == "__main__":
    root = Node(25)
    root.left = Node(15)
    root.right = Node(50)
    root.left.left = Node(10)
    root.left.right = Node(22)
    root.right.left = Node(35)
    root.right.right = Node(70)
    root.left.left.left = Node(4)
    root.left.left.right = Node(12)
    root.left.right.left = Node(18)
    root.left.right.right = Node(24)
    root.right.right.left = Node(66)
    root.right.right.right = Node(90)
    root.right.left.left = Node(31)
    root.right.left.right = Node(44)
 
    print("Inorder Traversal:",end=" ")
    printInorder(root)
    
    print("\nPostorder Traversal:",end=" ")
    printPostOrder(root)
    
    print("\nPreorder Traversal:",end=" ")
    printPreOrder(root)
    
```

```python
#Q2

class Graph:
    def __init__(self, n):
        adjlist = dict()

        for i in range(n):
            adjlist[f'{i+1}'] = []

        adjmat = []
        for _ in range(n):
            l = [0 for __ in range(n)]
            adjmat.append(l)
        
        self.adjlist = adjlist
        self.adjmat = adjmat
    
    def add_edge(self, u, v, weight):
        self.adjlist[f'{u}'].append([v, weight])
        self.adjmat[u-1][v-1] = weight
    
    def show(self):
        print('Adjacency List: ')
        print(self.adjlist)
        print('Adjacency Matrix: ')
        print(self.adjmat)


n = int(input('Enter the number of vertices: '))
edges = int(input('Enter the number of edges: '))

g = Graph(n)

print('Enter the edges of the graph in the format (u, v, weight): ')

for _ in range(edges):
    u, v, weight = tuple(map(int, input().split(' ')))
    g.add_edge(u, v, weight)

g.show()


```

```python
#Q3

from math import sqrt

def sort(arr):
    n = len(arr)
    
    for i in range(n-1):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def eucdist(x, y, n):
    dist = 0
    for i in range(n):
        dist = dist + (x[i]-y[i])**2
    
    return sqrt(dist)
    
def compute_dist(X, Y, n):
    points = len(X)
    dist = []
    
    for i in range(points):
        dist.append(eucdist(X[i], Y[i], n))
        
    return dist

n = int(input('Enter the dimensionality of points: '))
points = int(input('Enter the number of points: '))

print('*** Enter the points as - x0 x1 ... xn ***')

X = []
Y = []

print('--- List X ---')
for _ in range(points):
    print(f'Enter the value for x{_+1}: ')
    x = list(map(float, input().split()))
    X.append(x)
    
print('--- List Y ---')
for _ in range(points):
    print(f'Enter the value for y{_+1}: ')
    y = list(map(float, input().split()))
    Y.append(y)

dist = compute_dist(X, Y, n)

print('Distances Computed: ')
print(dist)

sort(dist)

print('Distances Sorted: ')
print(dist)

```
