```python
#Q1 Water Jug Problem using BFS(Breadth-First Search)

from collections import deque

def solveWaterJugBFS(jug1, jug2, target):
    visited = {}
    path = []
    isSolvable = False

    q = deque()
    q.append((0, 0))

    while len(q) > 0:
        u = q.popleft()

        if (u[0], u[1]) in visited:
            continue
        if (u[0] > jug1 or u[1] > jug2 or u[0] < 0 or u[1] < 0):
            continue

        path.append([u[0], u[1]])

        visited[(u[0], u[1])] = 1

        if u[0] == target or u[1] == target:
            isSolvable = True

            if u[0] == target:
                if u[1] != 0:
                    path.append([u[0], u[1]])
            else:
                if u[0] != 0:
                    path.append([0, u[1]])

            pathlen = len(path)

            print('(JUG 1, JUG 2) ')

            for i in range(pathlen):
                print("(", path[i][0], "  ,  ", path[i][1], ")")

            break

        q.append([u[0], jug2])
        q.append([jug1, u[1]])

        for x in range(max(jug1, jug2) + 1):
            c = u[0] + x
            d = u[1] - x

            if c == jug1 or (d == 0 and d >= 0):
                q.append([c, d])

            c = u[0] - x
            d = u[1] + x

            if (c == 0 and c >= 0) or d == jug2:
                q.append([c, d])

        q.append([jug1, 0])
        q.append([0, jug2])

    return isSolvable

def targetcheck(j1, j2, targetvalue):
    if targetvalue > j1 and targetvalue > j2:
        print("Invalid Input for Target")
        target1 = int(input("Enter input for target value again: "))
        return targetcheckconfirm(j1, j2, target1)
    else:
        return j1, j2, targetvalue

def targetcheckconfirm(ju1, ju2, targetvalue1):
    if targetvalue1 < ju2 or targetvalue1 < ju2:
        return ju1, ju2, targetvalue1
    else:
        return targetcheck(ju1, ju2, targetvalue1)

while True:
    j1 = int(input("Enter value for Jug 1: "))
    j2 = int(input("Enter value for Jug 2: "))
    target = int(input("Enter value for target: "))

    jugg1, jugg2, targett = targetcheck(j1, j2, target)

    solvable = solveWaterJugBFS(jugg1, jugg2, targett)

    if not solvable:
        print("No solution found.")
        retry = input("Retry? (y/n): ")
        if retry != "y":
            break
    else:
        break
```
