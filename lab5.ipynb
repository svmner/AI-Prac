```python
def create_weighted_directed_graph(edges):
    graph = defaultdict(dict)
    for start, end, weight in edges:
        graph[start][end] = weight
    return graph
  
edges = [['A', 'B', 6], ['A', 'C', 3], ['B', 'C', 1], ['C', 'B', 4], ['B', 'D', 2], ['C', 'D', 8], ['C', 'E', 2], ['D', 'E', 9], ['E', 'D', 7]]
graph = create_weighted_directed_graph(edges)


def dijkstra_algorithm(graph, start_node):
    queue = [start_node]
    shortest_distances = {node: float("inf") for node in graph}
    shortest_distances[start_node] = 0
    predecessor = {}

    while queue:
        current_node = queue.pop(0)
        for neighbor in graph[current_node]:
            potential_distance = shortest_distances[current_node] + graph[current_node][neighbor]
            if potential_distance < shortest_distances[neighbor]:
                shortest_distances[neighbor] = min(potential_distance, shortest_distances[neighbor])
                queue.append(neighbor)
                predecessor[neighbor] = current_node

    return shortest_distances, predecessor

def uniform_cost_search(graph, start_node, end_node):
    shortest_distances, predecessor = dijkstra_algorithm(graph, start_node)
    
    path = []
    current_node = end_node
    while current_node != start_node:
        if current_node not in predecessor:
            print("Path not reachable")
            break
        else:
            path.insert(0, current_node)
            current_node = predecessor[current_node]
    path.insert(0, start_node)
    
    if end_node in shortest_distances and shortest_distances[end_node] != float("inf"):
        print('Shortest distance is ' + str(shortest_distances[end_node]))
        print('And the path is ' + str(path))

uniform_cost_search(graph, 'A', 'D')
```
