```python
#DFS Q2

from collections import defaultdict

class Graph():
    def __init__(self,vertices):
        self.graph = defaultdict(list)
        self.V = vertices
        
    def Edge(self,u,v):
        self.graph[u].append(v)
        
    def Cyclic(self,v,visited,stack):
        visited[v] = True
        stack[v] = True
        
        for neighbour in self.graph[v]:
            if visited[neighbour] == False:
                if self.Cyclic(neighbour,visited,stack) == True:
                    return True
                elif stack[neighbour] == True:
                    return True
        
        stack = False
        return False 
    
    def checkCyclic(self):
        visited = [False]*(self.V+1)
        stack = [False]*(self.V+1)
        
        for node in range(self.V):
            if visited[node] == False:
                if self.Cyclic(node,visited,stack) == True:
                    return True
        return False
    
if __name__ == "__main__":
    g = Graph(4)
    g.Edge(0,1)
    g.Edge(0,2)
    g.Edge(2,0)
    g.Edge(1,2)
    g.Edge(2,3)
    g.Edge(3,3)
    
    if g.checkCyclic() == 1:
        print("Graph contains cycle")
    else:
        print("Graph does not contain cycle")
```

```python
#BFS Q2

from collections import defaultdict

class Graph:
    def __init__(self,vertices):
        self.graph=defaultdict(list)
        self.V=vertices

    def Edge(self,u,v):
        self.graph[u].append(v)

def CycleExist(n,graph):
    in_degree=[0]*n

    for i in range(n):
        for j in graph[i]:
            in_degree[j]+=1     
    queue=[]
   
    for i in range(len(in_degree)):
        if in_degree[i]==0:
            queue.append(i)
    cnt=0
    
    while(queue):
        nu=queue.pop(0)
        
        for v in graph[nu]:
            in_degree[v]-=1
            if in_degree[v]==0:
                queue.append(v)
        cnt+=1
        
    if cnt==n:
        return False
    else:
        return True
         
if __name__=='__main__':
    g=Graph(4)
    g.Edge(0,1)
    g.Edge(0,2)
    g.Edge(2,0)
    g.Edge(3,3)
    g.Edge(1,2)
     
    if CycleExist(g.V,g.graph):
        print("Graph contains cycle")
    else:
        print("Graph does not contain cycle")
```
